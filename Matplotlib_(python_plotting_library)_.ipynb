# Subplot in loop

```python
#subplot in for loop
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,6,3,4])
y=np.array([10,20,30,40])

for i in range(1,10):
  plt.subplot(3,3,i)
  plt.plot(x,y)
plt.show()
```

# To adjust the subplot spacing

```python
#to adjust the subplot spacing
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,6,3,4])
y=np.array([10,20,30,40])

for i in range(1,10):
  plt.subplot(3,3,i)
  plt.plot(x,y)

#to adjust the subplot spacing
plt.subplots_adjust(left=0.1,right=0.9,top=0.9,bottom=0.1,hspace=0.5,wspace=0.5)

plt.show()
```

# Nested pie chart

```python
outer_labels=["fruits","vegetables"]
outer_sizes=[60,40]

inner_labels=["apples","bananas","carrots","broccoli"]
inner_sizes=[30,30,20,20]

outer_colors=["skyblue","lightgreen"]
inner_colors=["lightcoral","orange","yellowgreen","mediumseagreen"]

#plotting the outer ring
fig,ax=plt.subplots()
ax.pie(outer_sizes,labels=outer_labels,colors=outer_colors,radius=1,wedgeprops=dict(width=0.3,edgecolor="w"))

#plotting the inner ring
ax.pie(inner_sizes,labels=inner_labels,colors=inner_colors,radius=0.7,wedgeprops=dict(width=0.3,edgecolor="w"))

plt.title("Nested pie chart example")
plt.show()
```
